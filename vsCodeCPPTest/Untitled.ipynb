```python
import numpy as np
from matplotlib import pyplot as plot
%matplotlib inline
from mpl_toolkits import mplot3d
plot.ion()
```

```python
data=np.loadtxt('mat.lab')[:,0:3]
label=np.loadtxt('mat.lab')[:,3]
```

```python
l=data.shape[0]
new_data=[]
for i in range(l):
    d=data[i:i+3,:]
    d=np.mean(d,axis=0)
    new_data.append(d)
data=np.array(new_data)
```

```python
fig = plot.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data[label==3,0],data[label==3,1],data[label==3,2])
ax.scatter(data[label==1,0],data[label==1,1],data[label==1,2])
ax.scatter(data[label==10,0],data[label==10,1],data[label==10,2])
ax.scatter(data[label==9,0],data[label==9,1],data[label==9,2])
ax.scatter(data[label==4,0],data[label==4,1],data[label==4,2])
ax.scatter(data[label==5,0],data[label==5,1],data[label==5,2])
ax.scatter(data[label==11,0],data[label==11,1],data[label==11,2])
ax.scatter(data[label==12,0],data[label==12,1],data[label==12,2])
```
